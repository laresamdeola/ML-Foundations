```python
import numpy as np
import matplotlib.pyplot as plt
import torch
import tensorflow as tf
import keras
```

```python
time = np.linspace(0, 100, 1000)
```

```python
# d = 2.5t

distance_robber = 2.5 * time
```

```python
# d = 3(t-5)

distance_sheriff = 3 * (time - 5)
```

```python
fig, ax = plt.subplots()
plt.title('Bank Robber Caught')
ax.plot(time, distance_robber, c='green')
ax.plot(time, distance_sheriff, c='purple')
ax.set_xlim([0, 40])
ax.set_ylim([0, 100])
plt.xlabel('time (in minutes)')
plt.ylabel('distance (in km)')
plt.axvline(x=30, c='orange', linestyle='--')
_ = plt.axhline(y=75, c='orange', linestyle='--')
plt.show()
```

```python
mark_I_energy = 1*time
mark_II_energy = 4 * (time - 30)
```

```python
fig, ax = plt.subplots()
plt.title('Solar Panel Energy Generation')
ax.plot(time, mark_I_energy, c='green')
ax.plot(time, mark_II_energy, c='purple')
ax.set_xlim([0, 100])
ax.set_ylim([0, 100])
plt.xlabel('days')
plt.ylabel('energy')
plt.axvline(x=40, c='orange', linestyle='--')
_ = plt.axhline(y=40, c='orange', linestyle='--')
plt.show()
```

```python
x_sample_1 = np.linspace(-10, 10, 1000)
```

```python
y1 = 3 * x_sample_1
```

```python
y2 = 1 + ((5 * x_sample_1)/2)
```

```python
fig, ax = plt.subplots()
plt.xlabel('x')
plt.ylabel('y')
ax.set_xlim([0, 3])
ax.set_ylim([0, 8])
ax.plot(x_sample_1, y1, c='r')
ax.plot(x_sample_1, y2, c='orange')
plt.axvline(x=2, c='purple', linestyle='--')
_ = plt.axhline(y=6, c='purple', linestyle='--')
```

```python
# x + y = 6
# 2x + 3y = 16
```

```python
# y3 = 6 - x
# y4 = 5.33 - 2x/3
```

```python
x_sample_2 = np.linspace(-10, 50, 1000)
```

```python
y3 = 6 - x_sample_2
y4 = 5.33 - ((2*x_sample_2)/3)
```

```python
fig, ax = plt.subplots()
plt.xlabel('x')
plt.ylabel('y')
ax.set_xlim([0, 4])
ax.set_ylim([0, 10])
ax.plot(x_sample_2, y3, c='lightblue')
ax.plot(x_sample_2, y4, c='coral')
plt.axvline(x=2, c='purple', linestyle='--')
_ = plt.axhline(y=4, c='purple', linestyle='--')
```
